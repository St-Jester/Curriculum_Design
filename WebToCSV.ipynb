# Data Science. Coursera. Parsing.

```python
browser = webdriver.Chrome()
base_url_target = "https://www.coursera.org"
url_target = 'https://www.coursera.org/browse/data-science/?facets=skillNameMultiTag%2CjobTitleMultiTag%2CdifficultyLevelTag%2Clanguages%2CentityTypeTag%3ACourses%2CpartnerMultiTag%2CcategoryMultiTag%3Adata-science%2CsubcategoryMultiTag&sortField='

browser.get(url_target)
links = []
```

```python
time.sleep(1)

elem = browser.find_element_by_tag_name("body")

no_of_pagedowns = 210

links = []
while no_of_pagedowns:
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.3)
    no_of_pagedowns-=1
    #this is for course name
#     post_elems = browser.find_elements_by_class_name("product-name")

    found_links = browser.find_elements_by_class_name("search-offering-card")

    for link in found_links:
         if link.get_attribute("href") not in links:
            links.append(link.get_attribute("href"))
print(len(links))
```

```python
all_dataDF = pd.DataFrame(links, 
             columns=['links'])
all_dataDF.drop_duplicates()
```

```python

all_dataDF.to_csv (r'C:\Users\stell\Documents\UCU\IntroToDS\export_dataframe.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
```

```python
all_data = []
i = 0;
for link in links:
    if(i%100 == 0):
        print("progress{}".format(i))
    i+=1;
    
    all_skils = []
     # Open link, read and parse content
    url = urllib.request.urlopen(link)
    content = url.read()
    soup = BeautifulSoup(content, 'lxml')
#happened to have to deal with the pages&infinite scroll pagination
    #coursename
    course_name = ''
    try:
        course_name = soup.find(class_="H2_1pmnvep-o_O-weightNormal_s9jwp5-o_O-fontHeadline_1uu0gyz max-text-width-xl m-b-1s").get_text()
    except AttributeError:
        course_name = 'N/A'
    #course rating
    course_rate = ''
    try:
        course_rate = soup.find(class_="H4_1k76nzj-o_O-weightBold_uvlhiv-o_O-bold_1byw3y2 m-l-1s m-r-1 m-b-0").get_text()
    except AttributeError:
        course_rate = 'N/A'
    #offered_by
    
    course_offeredby  = '';
    try:
        #todo make it be more universal, as classes are different, but image is still there
        course_offeredby = soup.find(class_="partnerBanner_np2ice-o_O-Box_120drhm-o_O-displayflex_poyjc").find('img')['title']
        
    except:
        try:
            course_offeredby = soup.find(class_="partnerBanner_np2ice-o_O-Box_120drhm-o_O-displayflex_poyjc").get_text()
        except:
            course_offeredby = 'N/A'
    
    #skills
    skils_data = soup.find('div',{'class': 'Box_120drhm-o_O-displayflex_poyjc-o_O-wrap_rmgg7w'})
    if skils_data:
        for skill in skils_data:
            item = skill.find('span', {'class' : 'centerContent_dqfu5r'})
            all_skils.append(item.get_text())
    else:
        all_skils.append(None)
        print(all_skils[-1:])
    
    all_data.append((course_name,course_rate,course_offeredby, all_skils))
```

```python
all_dataDF_final = pd.DataFrame(all_data, 
             columns=['name','rate','offered_by','skills'])

#data needs clearing
```

```python
frames = [all_dataDF_final, all_dataDF]
result = pd.concat(frames,axis=1, sort=False)
print(result.shape)
result.to_csv (r'..\full_data_coursera_datascience_courses_w_links.csv', index = None, header=False)
```
